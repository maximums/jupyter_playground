```python
plot1 = plt.subplot2grid((2, 4), (0, 0), colspan=2)
plot2 = plt.subplot2grid((2, 4), (0, 2), colspan=2)
plot3 = plt.subplot2grid((2, 4), (1, 0), colspan=4)
  
plot2.plot(a, a)
plot2.set_title('Utopie')
  
plot1.plot(b, c)
plot1.set_title('Haus')
  
plot3.plot(a, b)
plot3.set_title('Life')
  
plt.tight_layout()
plt.show()
```
